```python
#Get url from node label
def get_url_m(value):
    for key, val in labels_m.items():  # for name, age in dictionary.iteritems():  (for Python 2.x)
        if val == value:
            return key

        
def get_key(dictionary,value):
    for key, val in dictionary.items():
        if val == value:
            return key
```

```python
#Generate pathway for entire mechanism


def find_shortest_path(start_node,end_node,graph,df):
    
    node1 = str(identifiers[get_url_m(start_node)])
    node2 = str(identifiers[get_url_m(end_node)])
    #list of nodes
    path_nodes = g.get_shortest_paths(v=node1, to=node2, weights=None, mode='OUT')
    
    g_nodes = g.vs()['name'] 
    
    n1 = labels_m[get_key(identifiers,int(g_nodes[path_nodes[0][0]]))]
    for i in range(1,len(path_nodes[0])):
        idx = path_nodes[0][i]
        n2 = labels_m[get_key(identifiers,int(g_nodes[idx]))]
        for i in range(len(triples_list_m)):
            d = {}
            if n1 == labels_m[triples_list_m[i][0]] and n2 == labels_m[triples_list_m[i][2]]:
                d['S'] = labels_m[triples_list_m[i][0]]
                d['P'] = labels_m[triples_list_m[i][1]]
                d['O'] = labels_m[triples_list_m[i][2]]
                df = df.append(d,ignore_index=True)
        n1 = n2

    return df
    
```
